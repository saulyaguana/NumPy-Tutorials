## Agregar o eliminar dimensiones
#### Agregando con ndmin

```python
vector = np.array([1, 2, 3])
print(vector)
vector.ndim
```

```python
vector = np.array([1, 2, 3], ndmin=10)
print(vector)
vector.ndim
```

```python
expand = np.expand_dims(np.array([1, 2, 3]), axis=0)
print(expand)
expand.ndim
```

```python
expand = np.expand_dims(np.array([1, 2, 3]), axis=1)
print(expand)
expand.ndim
```

#### Eliminar dimensiones

+ Lo que hace np.squeeze() es reducir al número el dimensiones que no se esten usando

```python
print(vector, vector.ndim)
```

```python
vector2 = np.squeeze(vector)
print(vector2, vector2.ndim)
```
